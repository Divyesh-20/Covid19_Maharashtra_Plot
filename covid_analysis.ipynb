# India COVID-19 Data Analysis
This notebook performs basic data analysis using Pandas.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('india_covid19_data.csv')
df['Date'] = pd.to_datetime(df['Date'])

# Filter for Maharashtra
maharashtra_df = df[df['State'] == 'Maharashtra'].sort_values('Date')

# Print first few rows to check
print(maharashtra_df.head())

# Create the plot
plt.figure(figsize=(14, 6))

# Bar graph for daily confirmed cases
plt.bar(maharashtra_df['Date'], maharashtra_df['Confirmed'], color='lightblue', label='Daily Confirmed Cases')

# Line graph over the same data
plt.plot(maharashtra_df['Date'], maharashtra_df['Confirmed'], color='blue', marker='o', linewidth=1.5, label='Confirmed Cases Line')

# Plot styling
plt.title('Daily Confirmed COVID-19 Cases in Maharashtra', fontsize=14)
plt.xlabel('Date')
plt.ylabel('Confirmed Cases')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.legend()
plt.tight_layout()
plt.show()
```
